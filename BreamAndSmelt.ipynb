## 다향 회귀

```python
train_poly = np.column_stack((train_input**2, train_input)) # 농어의 무게 = a*길이^2 + b*길이+c 즉 곡선 형태로 만들기위하여 다향 회귀를 사용.
test_poly = np.column_stack((test_input**2, test_input))
print(train_poly.shape, test_poly.shape)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_poly)
train_scaled = ss.transform(train_poly)
test_scaled = ss.transform(test_poly) # 특성의 스케일을 정규화하는 과정.
```

## 릿지 회귀

```python
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge.fit(train_scaled, train_target)
print(ridge.score(train_scaled, train_target))
```

```python
print(ridge.score(test_scaled, test_target))
```

```python
import matplotlib.pyplot as plt
train_score = []
test_score = []
```

```python
alpha_list = [0.001, 0.01, 0.1, 1, 10, 100]
for alpha in alpha_list:
  ridge = Ridge(alpha=alpha)
  ridge.fit(train_scaled, train_target)
  train_score.append(ridge.score(train_scaled, train_target))
  test_score.append(ridge.score(test_scaled, test_target))
```

```python
plt.plot(np.log10(alpha_list), train_score)
plt.plot(np.log10(alpha_list), test_score)
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.show() # 오른쪽으로 갈수록 과소적합. 왼쪽은 테스트 점수가 너무 낮아 과대적합. -1, 즉 10정도가 적당함.
```

```python
ridge = Ridge(alpha=0.1)
ridge.fit(train_scaled, train_target)
print(ridge.score(train_scaled, train_target))
print(ridge.score(test_scaled, test_target))
```
